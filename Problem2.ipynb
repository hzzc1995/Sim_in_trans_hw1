```python
import numpy as np
import itertools

lamba = 5
radius = 50
velocity = 25
per_penalty_cost = 200
```

### 1. Expected distance traveled by the vehicle ###

```python
def TravelingSalsmanProblem(points):
    num_stops = len(points)
    possible_permutations = itertools.permutations(range(num_stops))
    dist_list = []
    for permutation_ in possible_permutations:
        dist_ = 0.0
        for i in range(len(permutation_)-1):
            dist_ = dist_ + GetDistance(points[permutation_[i]], points[permutation_[i+1]])
        dist_list.append(dist_)
    return np.min(dist_list)

def GetDistance(x, y):
    return np.sqrt( (x[0]-y[0])**2 + (x[1]-y[1])**2 )
```

```python
run = 10000
distance_list = []
for i in range(run):
    num_ = np.random.poisson(lam=lamba)
    coordinate_ = np.random.uniform(0, 2*np.pi, num_)
    points_ = [(radius*np.cos(angle_), radius*np.sin(angle_)) for angle_ in coordinate_]
    points_split_ = [points_[i:i+5] for i in range(0, len(points_), 5)]
    min_dist_sum_ = np.sum([TravelingSalsmanProblem(p_) for p_ in points_split_])
    min_dist_sum_ = min_dist_sum_ + radius * len(points_split_)
    distance_list.append(min_dist_sum_)
```

```python
print 'Expected distance: ', np.mean(distance_list)
```

### 3. Maximum amount of money to pay with penalty ###

```python
def TravelingSalsmanProblemWithPenalty(points):
    num_stops = len(points)
    possible_permutations = itertools.permutations(range(num_stops))
    dist_list = []
    optimal_dist, optimal_permutation = 10000, None
    for permutation_ in possible_permutations:
        dist_ = 0.0
        for i in range(len(permutation_)-1):
            dist_ = dist_ + GetDistance(points[permutation_[i]], points[permutation_[i+1]])
        dist_list.append(dist_)
        if dist_ < optimal_dist:
            optimal_dist = dist_
            optimal_permutation = permutation_
    
    dist_count = [radius]
    dist_count = [GetDistance(points[optimal_permutation[i]], points[optimal_permutation[i+1]]) for i in range(len(optimal_permutation)-1)]
    dist_count.append(radius)
    dist_count = np.cumsum(dist_count)
    
    return np.min(dist_list), dist_count
```

```python
run = 100
distance_list = []
penalty_list = []
for i in range(run):
    num_ = np.random.poisson(lam=lamba)
    coordinate_ = np.random.uniform(0, 2*np.pi, num_)
    points_ = [(radius*np.cos(angle_), radius*np.sin(angle_)) for angle_ in coordinate_]
    points_split_ = [points_[i:i+5] for i in range(0, len(points_), 5)]
    min_dist_sum_ = 0.0
    dist_count_tot_ = [0]
    for p_ in points_split_:
        dist_, count_ = TravelingSalsmanProblemWithPenalty(p_)
        min_dist_sum_ = min_dist_sum_ +  dist_
        min_dist_sum_ = min_dist_sum_ + radius * len(points_split_)
        dist_count_tot_.extend(count_+dist_count_tot_[-1])
    
    distance_list.append(min_dist_sum_)
    dist_count_tot_ = np.array(dist_count_tot_)
    num_penalty = dist_count_tot_[dist_count_tot_ > velocity * 8.0].shape[0]
    penalty_list.append(num_penalty)
```

```python
exp_cost = np.mean(penalty_list) * per_penalty_cost
print 'Expected penalty cost: ', exp_cost
```
